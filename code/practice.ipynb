```python
import pandas as pd
from iterstrat.ml_stratifiers import MultilabelStratifiedKFold
```

```python
train_data_path = '../data/feedback-prize-english-language-learning/train.csv'
df_train = pd.read_csv(train_data_path)
df_train
```

```python
skf = MultilabelStratifiedKFold(n_splits=10, shuffle=True, random_state=42)
target_cols = ['cohesion', 'syntax', 'vocabulary', 'phraseology', 'grammar', 'conventions']
for train_index, val_index in skf.split(df_train, df_train[target_cols]):
    print('len train', len(df_train.loc[train_index]))
    print('len val', len(df_train.loc[val_index]))
```

```python
result_train_data_path = '../data/feedback-prize-english-language-learning/custom_train.csv'
result_test_data_path = '../data/feedback-prize-english-language-learning/custom_test.csv'
df_train.loc[train_index].to_csv(result_train_data_path,index=False)
df_train.loc[val_index].to_csv(result_test_data_path,index=False)
```
